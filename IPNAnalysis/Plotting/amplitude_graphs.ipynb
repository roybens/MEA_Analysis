```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Load the data
df = pd.read_excel('/Users/mandarmp/Documents/RBS_LAB_DATA/quality_metrics_canine_organoids_90.xlsx')
```

```python
print(df.columns)
print(df.head())
```

```python
amplitude_median = df['amplitude_median']
```

```python
# Plotting the distribution of amplitude_median
plt.figure(figsize=(10, 6))
sns.histplot(amplitude_median, bins=50, kde=True, color='blue',
                stat='density')
plt.title('Distribution of Amplitude Median')
plt.xlabel('Amplitude Median')
plt.ylabel('Density')

plt.show()
```

```python
#get the summary statistics of the amplitude_median column
summary_stats = amplitude_median.describe()
print(summary_stats)
```

```python
amplitude_median_top_20 = df[df['amplitude_median']>summary_stats['75%']]
```

```python
# Plotting the distribution of amplitude_median_top_20
plt.figure(figsize=(10, 6))
sns.histplot(amplitude_median_top_20['amplitude_median'], bins=50
                , kde=True, color='blue', stat='density')
```

```python
summary_stats_top_25 = df[df['amplitude_median']>summary_stats['75%']]['amplitude_median'].describe()
print(summary_stats_top_25)
```

```python
df_after_4ap = pd.read_excel('/Users/mandarmp/Documents/RBS_LAB_DATA/quality_metrics_canine_organoids_94.xlsx')

amplitude_median_after_4ap = df_after_4ap['amplitude_median']

#plotting the distribution of amplitude_median_after_4ap
plt.figure(figsize=(10, 6))
sns.histplot(amplitude_median_after_4ap, bins=50, kde=True, color='blue',
                stat='density')
plt.title('Distribution of Amplitude Median After 4AP')
plt.xlabel('Amplitude Median After 4AP')
plt.ylabel('Density')
plt.show()

# describe the amplitude_median_after_4ap
summary_stats_after_4ap = amplitude_median_after_4ap.describe()
print(summary_stats_after_4ap)

#getting the top 20% of amplitude_median_after_4ap
amplitude_median_after_4ap_top_20 = df_after_4ap[df_after_4ap['amplitude_median']>summary_stats_after_4ap['75%']]
# Plotting the distribution of amplitude_median_after_4ap_top_20
plt.figure(figsize=(10, 6))
sns.histplot(amplitude_median_after_4ap_top_20['amplitude_median'], bins=50
                , kde=True, color='blue', stat='density')
plt.title('Distribution of Amplitude Median After 4AP Top 20%')
plt.xlabel('Amplitude Median After 4AP Top 25%')
plt.ylabel('Density')
plt.show()
# describe the amplitude_median_after_4ap_top_20
summary_stats_after_4ap_top_20 = df_after_4ap[df_after_4ap['amplitude_median']>summary_stats_after_4ap['75%']]['amplitude_median'].describe()
print(summary_stats_after_4ap_top_20)
```
