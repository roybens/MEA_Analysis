```python
def print_isi_ff_stats(stim_assay):
    isi = stim_assay.isi()
    isi_mean_std = stim_assay.calc_mean_isi(isi)
    fanofactors = stim_assay.calculate_fano_factor(isi_mean_std)
    print(f"Mean and std ISI: {isi_mean_std}")
    print(f"Fanofactors: {fanofactors}")
```
