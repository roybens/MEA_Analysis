```python
import numpy as np

def my_get_spike_cluster_times(tsne_results, spike_waveforms_dict):
    """
    Get spike times separated by cluster type from t-SNE analysis.

    Params:
        tsne_results: t-SNE results (numpy array of shape (n_samples, 2))
        spike_waveforms_dict: dictionary of spike waveforms (keys are spike times)

    Returns:
        dict : Dictionary containing lists of spike times (in seconds) for each cluster type
    """
    # Define cluster regions based on observed data
    regions = {
        'Arc Cluster': (tsne_results[:, 0] > -50) & (tsne_results[:, 0] < 40) & (tsne_results[:, 1] > -50) & (tsne_results[:, 1] < 50),
        'Top Right Cluster': (tsne_results[:, 0] > 40) & (tsne_results[:, 1] > 30),
        'Right Middle Cluster': (tsne_results[:, 0] > 40) & (tsne_results[:, 1] < 20) & (tsne_results[:, 1] > -20),
        'Bottom Right Cluster': (tsne_results[:, 0] > 40) & (tsne_results[:, 1] < -20),
        'Far Left Outliers': (tsne_results[:, 0] < -40),
        'Bottom Left Cluster': (tsne_results[:, 0] < -20) & (tsne_results[:, 1] < -40)
    }

    spike_times = np.array(list(spike_waveforms_dict.keys()))  # Convert spike times to NumPy array

    cluster_times = {}

    # Assign spike times to clusters
    for region_name, region_mask in regions.items():
        cluster_times[region_name] = spike_times[region_mask]

    return cluster_times



spike_times = my_get_spike_cluster_times(tsne_results, spike_waveforms)
```

```python
def my_plot_cluster_examples(spike_waveforms_dict, tsne_results, n_examples=3):
    """
    Plot example waveforms from different clusters to compare their shapes
    Params: 
        spike_waveforms_dict: dictionary of spike waveforms (from extract_spike_waveforms())
        tsne_results: t-SNE results (from tsne_spike_visualization())
        n_examples: number of examples to plot
    """
    # Convert dictionary to list for indexing
    waveforms = list(spike_waveforms_dict.values())
    
    # Define regions of interest based on your t-SNE plot
    regions = {
        'Arc Cluster': (tsne_results[:, 0] > -50) & (tsne_results[:, 0] < 40) & (tsne_results[:, 1] > -50) & (tsne_results[:, 1] < 50),
        'Top Right Cluster': (tsne_results[:, 0] > 40) & (tsne_results[:, 1] > 30),
        'Right Middle Cluster': (tsne_results[:, 0] > 40) & (tsne_results[:, 1] < 20) & (tsne_results[:, 1] > -20),
        'Bottom Right Cluster': (tsne_results[:, 0] > 40) & (tsne_results[:, 1] < -20),
        'Far Left Outliers': (tsne_results[:, 0] < -40),
        'Bottom Left Cluster': (tsne_results[:, 0] < -20) & (tsne_results[:, 1] < -40)
    }
    
    fig, axes = plt.subplots(len(regions), n_examples, figsize=(15, 10))
    
    for i, (name, mask) in enumerate(regions.items()):
        # Get indices for this region
        indices = np.where(mask)[0]
        
        # Randomly sample n_examples from this region
        sample_indices = np.random.choice(indices, size=min(n_examples, len(indices)), replace=False)
        
        for j, idx in enumerate(sample_indices):
            waveform = waveforms[idx]
            axes[i, j].plot(waveform)
            axes[i, j].set_title(f'{name}\nExample {j+1}')
            axes[i, 0].set_ylabel('Amplitude')
    
    plt.tight_layout()
    plt.show()


my_plot_cluster_examples(spike_waveforms, tsne_results)
```
