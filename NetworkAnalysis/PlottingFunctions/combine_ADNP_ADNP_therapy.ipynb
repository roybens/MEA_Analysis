```python
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Assuming df is your original DataFrame
columns_to_remove = ['cov_BurstDuration', 'mean_Burst_Peak_Abs', 'cov_Burst_Peak_Abs',
                     'mean_Burst_Peak', 'cov_Burst_Peak', 'Time', 'Chip_ID', 'Well', 'Run_ID',
                     'Chip_Well_Therapy_Key', 'Percentage_Change', 'Type']

div_values = [1, 3, 7]
therapy_types = ['GFP','CRISPRa',  'AAV']

for div in div_values:
    fig, axes = plt.subplots(1, 3, figsize=(24, 6))
    fig.suptitle(f'LDA Analysis for DIV {div}', fontsize=16)

    for idx, therapy in enumerate(therapy_types):
        # Filter the DataFrame for the current DIV and therapy type
        filtered_df = df[(df['DIV'] == div) & 
                         (df['Therapy_type'] == therapy) & 
                         (df['NeuronType'].isin(['WT', 'HET']))]
        
        # Remove specified columns and drop NaN values
        filtered_df = filtered_df.drop(columns=columns_to_remove).dropna()
        
        # Create a new column combining NeuronType and TherapyType
        filtered_df['CombinedType'] = filtered_df['NeuronType'] + '_' + filtered_df['Therapy_type']
        
        # Separate features and target variable
        X = filtered_df.drop(columns=['NeuronType', 'Therapy_type', 'CombinedType', 'DIV'])
        y = filtered_df['CombinedType']
        
        # Apply LDA
        lda = LinearDiscriminantAnalysis()
        X_lda = lda.fit_transform(X, y)
        
        # Create a DataFrame for LDA results
        lda_df = pd.DataFrame(X_lda, columns=['LD1'])
        lda_df['CombinedType'] = y.values
        
        # Plot the LDA results
        ax = axes[idx]
        sns.scatterplot(x='LD1', y=[0]*len(lda_df), hue='CombinedType', data=lda_df,
                        palette={f'WT_{therapy}': 'blue', f'HET_{therapy}': 'red'}, ax=ax)
        
        # Calculate the kernel density estimate
        wt_kde = stats.gaussian_kde(lda_df[lda_df['CombinedType'] == f'WT_{therapy}']['LD1'])
        het_kde = stats.gaussian_kde(lda_df[lda_df['CombinedType'] == f'HET_{therapy}']['LD1'])
        
        # Generate x values for plotting
        x_range = np.linspace(lda_df['LD1'].min(), lda_df['LD1'].max(), 100)
        
        # Calculate y values and normalize them to the range [0, 0.4]
        wt_y = wt_kde(x_range)
        het_y = het_kde(x_range)
        wt_y_normalized = wt_y / wt_y.max() * 0.4
        het_y_normalized = het_y / het_y.max() * 0.4
        
        # Plot the normalized density curves
        ax.fill_between(x_range, wt_y_normalized, alpha=0.3, color='blue')
        ax.fill_between(x_range, het_y_normalized, alpha=0.3, color='red')
        
        # Calculate means and standard deviations
        wt_data = lda_df[lda_df['CombinedType'] == f'WT_{therapy}']['LD1']
        het_data = lda_df[lda_df['CombinedType'] == f'HET_{therapy}']['LD1']
        
        mean_wt = np.mean(wt_data)
        mean_het = np.mean(het_data)
        std_wt = np.std(wt_data)
        std_het = np.std(het_data)
        
        # Plot vertical lines for means and add text annotations
        ax.axvline(mean_wt, color='blue', linestyle='--', linewidth=1)
        ax.axvline(mean_het, color='red', linestyle='--', linewidth=1)

        ax.text(mean_wt, 0.15, f'WT mean: {mean_wt:.2f}\nWT std: {std_wt:.2f}', 
                color='blue', ha='center', va='bottom', fontsize=8, rotation=90)
        ax.text(mean_het, 0.15, f'HET mean: {mean_het:.2f}\nHET std: {std_het:.2f}', 
                color='red', ha='center', va='bottom', fontsize=8, rotation=90)

        distance = mean_wt - mean_het

        # Plot horizontal line between the two means and add distance text
        ax.hlines(0.3, mean_het, mean_wt, colors='black', linestyles='-', linewidth=0.5)
        ax.text((mean_wt + mean_het) / 2, 0.32, f'Distance: {distance:.2f}', 
                color='black', ha='center', va='bottom', fontsize=8)
        
        ax.set_title(f'WT_{therapy} vs HET_{therapy}')
        ax.set_xlabel('Linear Discriminant 1')
        ax.set_ylabel('Linear Discriminant 2')
        ax.set_ylim([-0.2, 0.4])
        ax.legend(loc='best')

    plt.tight_layout()
   
    # Save the figure
    plt.savefig(f'/home/mmp/Documents/ADNP_JUNE2024/ADNP_therapy_differentvisualization//LDA_DIV{div}_All_Therapies.png', format='png', dpi=300, bbox_inches='tight')
    plt.close()  # Close the plot to free up memory

print("All graphs have been generated and saved.")
```
